```python
import jupyturtle as t
import math
t.make_turtle(delay=0.01, animate=True, height=300, width=540)

def draw_retangle(x,y,width,height):
    t.jump_to(x,y)
    for i in range(4):
        if i%2 == 0:
            t.fd(width)       
        else:
            t.fd(height)
        t.rt(90)        
def draw_pentagram(x,y,radiu):
    length =2*radiu*math.sin(60*3.1415/180)
    y_ = y-length/2*math.tan(18*3.1415/180)
    x_ = x-length/2
    t.jump_to(x_,y_)
    for i in range(5):
        t.fd(length)
        t.rt(144)
    t.jump_to(x,y)

draw_retangle(10,10,520,250)    

x=30
y=30
length = 10
for i in range(5):
    if i%2 == 0:
        for j in range(10):
            draw_pentagram(x+50*j,y+50*i,10)
    else:
        for k in range(10):
            draw_pentagram(x+50*k+30,y+50*i,10)
```

##### with读txt文件 可不用file.close()  

 - r 打开只读文件，该文件必须存在。
 - r+ 打开可读写的文件，该文件必须存在。
 - w 打开只写文件，若文件存在则文件长度清为0，即该文件内容会消失。若文件不存在则建立该文件。
 - w+ 打开可读写文件，若文件存在则文件长度清为零，即该文件内容会消失。若文件不存在则建立该文件。
 - a 以附加的方式打开只写文件。若文件不存在，则会建立该文件，如果文件存在，写入的数据会被加到文件尾，即文件原先的内容会被保留。
 - a+ 以附加方式打开可读写的文件。若文件不存在，则会建立该文件，如果文件存在，写入的数据会被加到文件尾后，即文件原先的内容会被保留。


```python
with open("test.txt","a") as file:
    file.write("1,3,5,7,9\n")
    file.write("2,3,4")
```

```python
import random

i = 1
str_ = ""
while i<=100:
    a = random.uniform(0,100)
    #print('{:08.4f}'.format(a),end=',')
    b = '{:08.4f},'.format(a)
    str_ += b
    if i %10 ==0:
        str_ += "\n"
    i +=1
list_ = str_.split("\n")

with open("homework4.txt","w") as file:
    for i in list_:
        file.write(i[:-2])
        file.write("\n")

with open("homework4.txt","r") as file:
    content = file.read()
    print(content)
```
