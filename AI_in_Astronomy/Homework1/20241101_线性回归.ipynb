```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
from scipy.stats import pearsonr
```

```python
# 数据可视化
X_data = np.array([0.21, 0.26, 0.27, 0.27, 0.45, 0.5, 0.8, 1.1, 1.4, 2.0])
y_data = np.array([130, 70, 185, 220, 200, 270, 300, 450, 500, 800])
plt.scatter(X_data, y_data)
plt.show()
```

```python
X_data = np.arange(60)

Y_data = []
for i in X_data:
    y = 0.5*i+20
    n=np.random.uniform(-1,1)*2
    y = y+n
    Y_data.append(y)

Y_data = np.array(Y_data)
plt.scatter(X_data, Y_data)
plt.show()
```

```python
np.corrcoef(X_data, Y_data)
```

```python
from scipy import stats

res = stats.linregress(X_data, Y_data)

y_data = res.intercept + res.slope * X_data
```

```python
plt.scatter(X_data,Y_data,color="blue")
plt.plot(X_data,y_data,color="red")
print(f"y=20 + 0.5* x")
print(f"y={res.intercept} + {res.slope} * x")
```

```python
from sklearn.linear_model import LinearRegression

X_train = np.array(X_data).reshape(-1, 1)
Y_train = np.array(Y_data)
```

```python
# 构建模型
regr = LinearRegression()
regr.fit(X_train, Y_train)
```

```python
import matplotlib
# 指定默认字体  
matplotlib.rcParams['font.family']='sans-serif'  
matplotlib.rcParams['font.sans-serif'] = ['SimHei']   
#解决负号'-'显示为方块的问题  
matplotlib.rcParams['axes.unicode_minus'] = False 
```
