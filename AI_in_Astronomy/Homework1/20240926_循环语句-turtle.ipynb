```python
import jupyturtle as t
import math

def circle(radius, steps=360):
    angle = 360 / steps
    length = 2 * math.pi * radius / steps
    for _ in range(steps):
        t.forward(length)
        t.right(angle)
def backward(length):
    t.forward(-length)
    
def tree1(length):
    if length <= 0:
        return
    else:
        t.forward(length)
        t.left(30)
        tree1(length - 10)
        t.right(60)
        tree1(length - 10)
        if length-10<=0:
            t.set_color("pink")
            circle(5)
            t.set_color("brown")
        t.left(30)
        backward(length)
t.make_turtle(delay=0.0001,width=800,height=800)
t.right(90)
t.forward(100)
t.left(180)
tree1(100)
##实在太久了，跑了好久还没结束，画个半树
```
