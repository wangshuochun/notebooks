[Pandas](http://pandas.pydata.org/)是一个基于 Numpy 的 Python 数据分析库。Pandas 提供了高效易用的数据结构与数据分析工具，应用领域广泛，主要特点如下：
- 快速高效的 DataFrame 对象，非常适合于 SQL 数据库或 Excel 中的表格类数据。
- 能够处理有序和无序（不一定是等间隔）的时间序列数据。
- 支持带有行和列标签的任意的矩阵数据（同构或异构数据类型）。
- 快捷地把其它形式的观测或统计数据导入到Pandas数据结构中。
- 灵活处理缺失数据

#### 字典转化为 series

```python
times = np.array([6, 7, 8, 11, 12, 13, 18, 19, 20, 21])
customers = np.array([60, 70, 65, 100, 230, 150, 100, 300, 250, 150])
costs = np.array([6, 7, 8, 24, 23, 26, 45, 55, 45, 40])
```

```python
data = {"customers":customers,"costs":costs}
df = pd.DataFrame(data,index = times)
df
```

## 生成一个实验数据

一个班有40个人，每个人考4门，语文（90，rms=20），数学（120，rms = 30），英语（50，rms = 20），政治（80+-15）

```python
data1 =np.random.normal(90,20,40).astype(int)
data2 =np.random.normal(120,30,40).astype(int)
data3 =np.random.normal(50,20,40).astype(int)
data4 =np.random.normal(80,15,40).astype(int)
```

```python
index_ = np.array(40)
data ={"语文":data1,"数学":data1,"英语":data3,"政治":data4}

df = pd.DataFrame(data)
```

```python
first_name = ["王", "李", "张", "刘", "赵", "蒋", "孟", "陈", "徐", "杨", "沈", "马", "高", "殷", "上官", "钟", "常"]
second_name = ["伟", "华", "建国", "洋", "刚", "万里", "爱民", "牧", "陆", "路", "昕", "鑫", "兵", "硕", "志宏", "峰", "磊", "雷", "文","明浩", "光", "超", "军", "达"]

def name(name):
    first = first_name[np.random.randint(low=0, high=17)]
    second = second_name[np.random.randint(low=0, high=24)]
    name = first + second
    return name
    
n=[]
for i in range(40):
    namei = name(name)
    n.append(namei)
print(n)
                        
```

```python
df = pd.DataFrame(data,index=n)
```
